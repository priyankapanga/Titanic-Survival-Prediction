This is a project exploring the Titanic data set, comparing the accuract of two machine learning models: Random Forest and Logistic Regression. 

Importing the libraries: 

```python

import numpy as np 
import pandas as pd 


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data =pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
```

First we can use the RandomForest model: 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd

y = train_data["Survived"]
features = ["Pclass", "Sex", "SibSp", "Parch"]


X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

# Splitting the training data into training and validation sets-> to calculate the accuracy
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y_train)

predictions = model.predict(X_val)

accuracy = accuracy_score(y_val, predictions)

print(f"Accuracy: {accuracy * 100:.2f}%")

# Predictions on the test data
test_predictions = model.predict(X_test)

# saving output to csv file( used in submission to Kaggle's competition)
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': test_predictions})
output.to_csv('randomforest_output.csv', index=False)
print("Your submission was successfully saved!")
```

Now, comparing this accuracy rate to the Logistic Regression model: 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd

y = train_data["Survived"]
features = ["Pclass", "Sex", "SibSp", "Parch"]


X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)

#using logistic regression
model = LogisticRegression(random_state=1)
model.fit(X_train, y_train)

# Make predictions on validation set
predictions = model.predict(X_val)

# Calculating the accuracy
accuracy = accuracy_score(y_val, predictions)
print(f"Accuracy: {accuracy * 100:.2f}%")


test_predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': test_predictions})
output.to_csv('regression_output.csv', index=False)

print("Your submission was successfully saved!")
```

The logistic regression model showed a higher accuracy rate than the random forest model, by 2.3%.
